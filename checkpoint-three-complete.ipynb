```python
#I wonder if a histogram could be more descriptive for the ones missing a lot of data?
#Not really, the X axis says index and lists numbers but would be easier to understand if I knew what the data referred to.

for col in public.columns:
    missing = public[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:  
        print('created missing indicator for: {}'.format(col))
        public['{}_ismissing'.format(col)] = missing

ismissing_cols = [col for col in public.columns if 'ismissing' in col]
public['num_missing'] = public[ismissing_cols].sum(axis=1)

public['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y='num_missing')
```
